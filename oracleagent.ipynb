```python
import seaborn as sns
import matplotlib.pyplot as plt
import pickle
import matplotlib.cm as cm

def plot_rewards_subplots(env_name, m, n, rewards_ddpg, rewards_adv_ddpg_good_kl, rewards_adv_kl_good_ddpg):
    x = [(i+1)*1000 for i in range(10)]

    fig, axs = plt.subplots(1, 3, figsize=(18, 8))  # Create a new figure with 3 subplots

    colors = cm.tab10(np.arange(10)/9.)  # Define colors for different agents

    rewards = [rewards_ddpg, rewards_adv_ddpg_good_kl, rewards_adv_kl_good_ddpg]
    titles = ['DDPG VS DDPG', 'Adv: DDPG VS Good: OracleKL', 'Adv: OracleKL VS Good: DDPG']

    for r in range(3):
        for i in range(m+n):
            if i < m:  # Plot Adv agent's rewards
                axs[r].plot(x, rewards[r][i*10:(i+1)*10], color=colors[i % len(colors)], label=f"Adv Agent {i+1}")
            else:  # Plot Good agent's rewards
                j = i - m
                axs[r].plot(x, rewards[r][i*10:(i+1)*10], color=colors[i % len(colors)], label=f"Good Agent {j+1}")

        axs[r].set_title(f"{env_name}: {titles[r]}")  # Add a title to the subplot
        axs[r].set_xlabel("Time steps")  # Add x-axis label
        axs[r].set_ylabel("Rewards")  # Add y-axis label
        axs[r].legend()  # Show legend

    plt.tight_layout()  # Ensure subplots do not overlap
    plt.savefig(env_name+'.eps')
    plt.show()  # Display the plot
```

```python
def plot_reward_differences(env_name, m, n, rewards_ddpg, rewards_adv_ddpg_good_kl, rewards_adv_kl_good_ddpg, save_path):
    plt.figure(figsize=(20,10))

    # Compute differences
    diff_adv = np.array(rewards_adv_kl_good_ddpg) - np.array(rewards_adv_ddpg_good_kl)
    diff_good = np.array(rewards_adv_kl_good_ddpg) - np.array(rewards_ddpg)

    for i in range(m+n):
        if i < m:
            label = 'Adv Agent'
        else:
            label = 'Good Agent'

        plt.subplot(2, 1, 1)
        plt.plot(diff_adv[i*10:(i+1)*10], label=f'{label} {i+1}')
        plt.title(f"[{env_name}]: Difference in rewards for Adv Agents (KL-DDPG)")
        plt.xlabel("Iteration Number")
        plt.ylabel("Difference in Reward")
        plt.legend()

        plt.subplot(2, 1, 2)
        plt.plot(diff_good[i*10:(i+1)*10], label=f'{label} {i+1}')
        plt.title(f"[{env_name}]: Difference in rewards for Good Agents (KL-DDPG)")
        plt.xlabel("Iteration Number")
        plt.ylabel("Difference in Reward")
        plt.legend()

    plt.tight_layout()  # Ensure subplots do not overlap
    plt.savefig(save_path)  # Save the plot to your Google Drive
    plt.show()  # Display the plot
```
